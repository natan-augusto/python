## 1. Functions
These are reusable blocks of code that perform a specific task.
Some examples of functions in Python:

### Function: print()
The print function prints the data on the terminal.
To enter string data (text) into a print function, we use double quotes (") or single quotes ('), depending on whether the text contains single or double quotes.  
Example:

```print("You're welcome!")```: Example of the use of double quotation marks.  
```print('"Hello, World"')```: Example of the use of single quotes.

You can group separate data items in a print statement using a comma.  
Example:  

```print("Hello,", "World")```: Example of using a comma as a separator for arguments (data).

````python
#Function: print()
# The print function prints the data on the terminal.
print("Hello, World")
print() # Prints a blank line.
print("You're welcome!")
print('"Hello, World"')
print()
print("Hello,", "World") 
````

### Function: input()
The input() function collects user-entered data, which can then be assigned to a variable.

````python
#Function: input()
# The input function collects user data, which can then be assigned to a variable.
name = input("What's your name? ")
print("Hello,", name)

# However, variables must be converted beforehand if they are not of the string type (text) using the input() function; otherwise, the input() function will always return the input as a string (text).
# int(): Function for converting to integers.
age = int(input("How old are you? "))
print("You are", age, "years old.")
# float(): Function to convert to rational numbers
weight = float(input("How much do you weigh? "))
print("You weigh", weight, "kg.")
````
